```python
URL = 'https://www.respublica.ru/authors/15753'
HEADERS ={}
HOST = 'https://www.respublica.ru'
```

```python
def get_html(url, params = None):
    r=requests.get(url, params = params)
    return r

def next_page(html):
    soup=BeautifulSoup(get_html(URL).text, 'html.parser')
    next_page = soup.find_all('a', class_="pagination-next")
    if next_page:
        return True
    else:
        return False
    
def get_pages_count(html):
    soup=BeautifulSoup(get_html(URL).text, 'html.parser')
    next_page = soup.find_all('a', class_="pagination-next")
    if next_page:
        return HOST+next_page[0].get('href')
    else:
        return

def get_content(html):
    soup=BeautifulSoup(html, 'html.parser')
    items = soup.find_all('div', class_="rd-listing-product-item rd-listing-product-item_size- rd-listing-product-item_type-"  )
    links=[]
    for item in items:
        links.append(HOST+item.find('a',class_="rd-listing-product-item__image-wrapper").get('href'))
    print(links)
    return links
    
def all_pages(URL):
    URLs = [URL]
    while True:
        try:
            soup=BeautifulSoup(get_html(URL).text, 'html.parser')
            next_page = soup.find_all('a', class_="pagination-next")
            URL = HOST+next_page[0].get('href')
            print(URL)
            URLs.append(URL)
        except:
            print('end')
            break
    return URLs

def all_cont(URL):
    all_cont=[]
    for u in all_pages(URL):
        all_cont.extend(get_content(get_html(u).text))
    return all_cont
```

```python
dict_auth=dict()
for a in authors:
    URL='https://www.respublica.ru/authors/'+str(a)
    dict_auth[a]=all_cont(URL)
```
